Passando Selic para DF

```python
df_selic = pd.DataFrame(selic)
```

```python
df_selic.sample()
```

```python
df_selic['data'] = pd.to_datetime(df_selic['data'], format='%d/%m/%Y')
```

```python
df_selic.set_index('data', inplace=True)
# Resample para pegar o primeiro valor de cada ms
df_selic_mensal = df_selic.resample('M').first()
# Formatando a data para o formato desejado (Ex: jun/2016)
df_selic_mensal.index = df_selic_mensal.index.strftime('%b/%Y')
```

```python
df_selic_mensal.tail(12)
```
