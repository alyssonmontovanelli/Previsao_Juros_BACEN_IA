- Criando DF para cada métrica obtida através de requisição

```python
def cria_dataframe(data, nome_coluna):
   df = pd.DataFrame(data)
   df['data'] = pd.to_datetime(df['data'], format='%d/%m/%Y')
   df.set_index('data', inplace=True)
   df.rename(columns={"valor": nome_coluna}, inplace= True)
   df_valor_mensal = df.resample('M').first()
   df_valor_mensal.index = df_valor_mensal.index.strftime('%b/%Y')
   df_valor_mensal.index = pd.to_datetime(df_valor_mensal.index, format='%b/%Y')
   df_selic_filtrado = df_valor_mensal[df_valor_mensal.index >= pd.to_datetime('2002-01-01')]

   return df_selic_filtrado
```
